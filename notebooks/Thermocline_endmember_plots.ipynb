# max and min of endmember parameter ranges (used to calculate the parameter weights)

```python
max(SPSTSW_df["silicate"])
```

```python
dataframes = [SPSTSW_df, SPCW_df, ESSW_df, ENPCW_df, PSUW_df]
parameters = ["conservative_temp", "absolute_salinity", "oxygen", "silicate", "nitrate", "phosphate"]
temp_max = 0
temp_min = 100
sal_max = 0
sal_min = 100
oxy_max = 0
oxy_min = 100
si_max = 0
si_min = 100
n_max = 0
n_min = 100
p_max = 0
p_min = 100

print("Calculating maximum and minimum of parameters across endmembers in the thermocline")
for df in dataframes:
  for param in parameters:
    if param == "conservative_temp"and max(df[param])> temp_max:
      temp_max=max(df[param])
    elif param == "absolute_salinity"and max(df[param])> sal_max:
      sal_max=max(df[param])
    elif param == "oxygen"and max(df[param])> oxy_max:
      oxy_max=max(df[param]) 
    elif param == "silicate"and max(df[param])> si_max:
      si_max=max(df[param])
    elif param == "nitrate"and max(df[param])> n_max:
      n_max=max(df[param]) 
    elif param == "phosphate"and max(df[param])> p_max:
      p_max=max(df[param]) 
    elif param == "conservative_temp"and min(df[param])< temp_min:
      temp_min=min(df[param])
    elif param == "absolute_salinity"and min(df[param])< sal_min:
      sal_min=min(df[param])
    elif param == "oxygen"and min(df[param])< oxy_min:
      oxy_min=min(df[param]) 
    elif param == "silicate"and min(df[param])< si_min:
      si_min=min(df[param])
    elif param == "nitrate"and min(df[param])< n_min:
      n_min=min(df[param])
    elif param == "phosphate"and min(df[param])< p_min:
      p_min=min(df[param])  
print(temp_max)
print(temp_min)
print(sal_max)
print(sal_min)
print(oxy_max)
print(oxy_min)
print(si_max)
print(si_min)
print(n_max)
print(n_min)
print(p_max)
print(p_min)     
```

```python
print("Below are the ranges for each parameter in the thermocline.")
print("temperature", temp_max-temp_min)
print("salinity", sal_max-sal_min)
print("oxygen", oxy_max-oxy_min)
print("silicate", si_max-si_min)
print("nitrate", n_max-n_min)
print("phosphate", p_max-p_min)
```

```python
print("Below is 100 divided by ranges for each parameter in the thermocline to give each paramter the same weight based on their range.")
print("temperature", 100/(temp_max-temp_min))
print("salinity", 100/(sal_max-sal_min))
print("oxygen", 100/(oxy_max-oxy_min))
print("silicate", 100/(si_max-si_min))
print("nitrate", 100/(n_max-n_min))
print("phosphate", 100/(p_max-p_min))
```
